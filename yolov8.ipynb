```python
import os
import pandas as pd
import base64

# 1. Recreate your metrics table
model_stats = {
    "Model": ["YOLOv8 Nano (Custom)"],
    "mAP@50": [0.905],
    "mAP@50-95": [0.728],
    "Inference (ms)": [3.76],
    "FPS": [265.7],
    "Parameters (M)": [3.01],
    "GFLOPs": [8.1]
}
df = pd.DataFrame(model_stats)
html_table = df.to_html(classes='table table-striped', index=False)

# 2. Function to embed images directly into the HTML
def get_base64_image(image_path):
    if os.path.exists(image_path):
        with open(image_path, "rb") as img_file:
            return base64.b64encode(img_file.read()).decode('utf-8')
    return None

# List the specific graphs you want in the report
graphs_to_include = [
    'confusion_matrix.png',
    'PR_curve.png',
    'F1_curve.png',
    'results.png' # This contains the loss curves over all 50 epochs
]

base_path = '/kaggle/working/runs/rod_yolov8_training/'
images_html = ""

for graph in graphs_to_include:
    img_b64 = get_base64_image(base_path + graph)
    if img_b64:
        images_html += f"""
        <div style="margin-bottom: 30px;">
            <h3>{graph.replace('.png', '').replace('_', ' ').title()}</h3>
            <img src="data:image/png;base64,{img_b64}" style="max-width: 100%; height: auto; border: 1px solid #ddd;">
        </div>
        """

# 3. Build the final HTML document
html_template = f"""
<!DOCTYPE html>
<html>
<head>
    <title>YOLOv8 Obstacle Detection Report</title>
    <style>
        body {{ font-family: Arial, sans-serif; margin: 40px; color: #333; }}
        h1 {{ color: #2c3e50; }}
        .table {{ width: 100%; border-collapse: collapse; margin-bottom: 40px; }}
        .table th, .table td {{ padding: 12px; border: 1px solid #ddd; text-align: center; }}
        .table th {{ background-color: #f4f4f4; }}
    </style>
</head>
<body>
    <h1>Model Evaluation Report: ROD Dataset</h1>
    
    <h2>1. Core Metrics</h2>
    {html_table}
    
    <h2>2. Visual Analytics</h2>
    {images_html}
    
</body>
</html>
"""

# 4. Save the document
with open('/kaggle/working/final_model_report.html', 'w') as f:
    f.write(html_template)

print("Report generated successfully! Download 'final_model_report.html' from your output folder.")
```

```python
# 1. Install reportlab if it isn't already in the environment
!pip install -q reportlab

import os
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle

# 2. Paths to your YOLOv8 assets (Adjust run_dir if your folder name is different)
run_dir = '/kaggle/working/runs/rod_yolov8_training'
pdf_path = '/kaggle/working/YOLOv8_Obstacle_Detection_Report.pdf'

confusion_matrix_path = f"{run_dir}/confusion_matrix.png"
pr_curve_path = f"{run_dir}/PR_curve.png"

# 3. Initialize PDF document setup
doc = SimpleDocTemplate(pdf_path, pagesize=letter,
                        rightMargin=40, leftMargin=40, topMargin=40, bottomMargin=40)
story = []
styles = getSampleStyleSheet()

# Create custom styles for academic presentation
title_style = ParagraphStyle(
    'DocTitle',
    parent=styles['Heading1'],
    fontSize=22,
    leading=26,
    textColor=colors.HexColor("#2C3E50"),
    spaceAfter=15
)

h2_style = ParagraphStyle(
    'SectionHeader',
    parent=styles['Heading2'],
    fontSize=14,
    leading=18,
    textColor=colors.HexColor("#2C3E50"),
    spaceBefore=12,
    spaceAfter=6,
    keepWithNext=True
)

body_style = ParagraphStyle(
    'BodyTextCustom',
    parent=styles['BodyText'],
    fontSize=10,
    leading=14,
    textColor=colors.HexColor("#333333"),
    spaceAfter=8
)

callout_style = ParagraphStyle(
    'CalloutText',
    parent=styles['BodyText'],
    fontSize=10,
    leading=14,
    textColor=colors.HexColor("#1B4F72"),
    spaceAfter=8
)

# --- Content Generation ---

# Title & Objective
story.append(Paragraph("Model Evaluation Report: YOLOv8 Nano", title_style))
story.append(Paragraph("<b>Objective:</b> To train and evaluate a highly efficient, real-time obstacle detection model optimized for low-latency deployment on edge hardware and autonomous systems.", body_style))
story.append(Spacer(1, 10))

# Section 1: Core Metrics Table
story.append(Paragraph("1. Executive Summary & Core Metrics", h2_style))
story.append(Paragraph("The model was trained for 50 epochs using the Ultralytics YOLOv8 Nano architecture. It successfully balances localization accuracy with low computational overhead.", body_style))

metrics_data = [
    [Paragraph("<b>Metric</b>", body_style), Paragraph("<b>Value / Score</b>", body_style), Paragraph("<b>Operational Significance</b>", body_style)],
    [Paragraph("mAP@50", body_style), Paragraph("90.5%", body_style), Paragraph("General intersection accuracy at a 50% threshold.", body_style)],
    [Paragraph("mAP@50-95", body_style), Paragraph("72.8%", body_style), Paragraph("Strict academic standard for high localization precision.", body_style)],
    [Paragraph("Inference Time", body_style), Paragraph("3.8 ms", body_style), Paragraph("Pure computational latency per frame.", body_style)],
    [Paragraph("Throughput (FPS)", body_style), Paragraph("265.7 FPS", body_style), Paragraph("Capable of real-time stream processing.", body_style)],
    [Paragraph("Parameters", body_style), Paragraph("3.01 M", body_style), Paragraph("Extremely lightweight footprint.", body_style)],
    [Paragraph("GFLOPs", body_style), Paragraph("8.1", body_style), Paragraph("Low battery consumption on mobile hardware.", body_style)]
]

t = Table(metrics_data, colWidths=[100, 80, 320])
t.setStyle(TableStyle([
    ('BACKGROUND', (0,0), (-1,0), colors.HexColor("#EAECEE")),
    ('ALIGN', (0,0), (-1,-1), 'LEFT'),
    ('VALIGN', (0,0), (-1,-1), 'TOP'),
    ('TEXTCOLOR', (0,0), (-1,0), colors.HexColor("#2C3E50")),
    ('BOTTOMPADDING', (0,0), (-1,-1), 6),
    ('TOPPADDING', (0,0), (-1,-1), 6),
    ('GRID', (0,0), (-1,-1), 0.5, colors.HexColor("#BDC3C7")),
]))
story.append(t)
story.append(Spacer(1, 15))

# Section 2: Accuracy Analysis & PR Curve
story.append(Paragraph("2. Accuracy & Bounding Box Localization Analysis", h2_style))
story.append(Paragraph("An mAP@50 score of 90.5% indicates that when the network predicts a bounding box around an obstacle, it successfully overlaps the actual ground-truth target by at least 50% in more than 9 out of 10 cases. The strict mAP@50-95 score of 72.8% serves as the baseline proof that the model draws highly stable and closely fitted boxes, minimizing false detection areas.", body_style))

if os.path.exists(pr_curve_path):
    story.append(Paragraph("<b>Precision-Recall (PR) Curve:</b>", body_style))
    img_pr = Image(pr_curve_path, width=320, height=240)
    img_pr.hAlign = 'CENTER'
    story.append(img_pr)
story.append(Spacer(1, 15))

# Section 3: Confusion Matrix & Class Insights
story.append(Paragraph("3. Class-Level Performance & Matrix Anomalies", h2_style))
story.append(Paragraph("The network performs cleanly across most structural categories. Rigid obstacles featuring uniform geometries and unique chromatic signatures—such as Fire Hydrants (99.5% mAP50), Electrical Boxes (99.5% mAP50), and Traffic Cones (97.6% mAP50)—achieved near-perfect convergence.", body_style))
story.append(Paragraph("Conversely, continuous horizontal surface bounds like 'Road' (77.2% mAP50) and highly scale-variable targets like 'Traffic Signs' (69.9% mAP50) show slightly lower rates due to standard square box limitations over angled perspective gradients.", body_style))

if os.path.exists(confusion_matrix_path):
    story.append(Paragraph("<b>Confusion Matrix Graph:</b>", body_style))
    img_cm = Image(confusion_matrix_path, width=300, height=240)
    img_cm.hAlign = 'CENTER'
    story.append(img_cm)
story.append(Spacer(1, 15))

# Section 4: Deployment Viability
story.append(Paragraph("4. Edge Deployment Viability & Execution Summary", h2_style))
story.append(Paragraph("<b>Hardware Target:</b> Tesla T4 GPU Baseline / Edge Microcontrollers", callout_style))
story.append(Paragraph("With a mathematical volume constrained to 3.01 Million parameters and a computational profile requiring only 8.1 GFLOPs, the network bypasses the heavy memory constraints typical of deeper structural detectors. The complete 3.8 ms inference latency ensures the model handles high frame rates without hitting memory bottlenecks or inducing thermal throttling on target platforms.", body_style))

# Build PDF
doc.build(story)
print(f"PDF Successfully generated and saved to: {pdf_path}")
```
